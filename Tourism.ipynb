```python
!pip install -q pandas numpy scikit-learn matplotlib tensorflow plotly

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import os

datasetPath = '/content/drive/MyDrive/Dataset'

tourismRating = pd.read_csv(os.path.join(datasetPath,'tourism_rating.csv'))
tourismPlaces = pd.read_csv(os.path.join(datasetPath,'tourism_with_id.csv'))
userData = pd.read_csv(os.path.join(datasetPath,'user.csv'))

print('tourism rating :', tourismRating.shape, '| columns:', list(tourismRating.columns))
print('tourism places :', tourismPlaces.shape, '| columns:', list(tourismPlaces.columns))
print('user data :', userData.shape, '| columns:', list(userData.columns))

df = tourismRating.merge(
    tourismPlaces[['Place_Id','Place_Name','Category','City']],
    on='Place_Id', how='left'
)

print('\nMerged dataframe :', df.shape)
print(df.head())

colUser='User_Id'
colPlace='Place_Id'
colRating='Place_Ratings'
colName='Place_Name'
colCategory='Category'
colLocation='City'

epochs=40
embeddingDim=64
batchSize=256
learningRate=0.001
validationSplit=0.15
testSplit=0.10
topN=10

import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,Model,regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

import warnings,pickle
warnings.filterwarnings('ignore')

tf.random.set_seed(42)
np.random.seed(42)

print("TensorFlow :", tf.__version__)
print("GPU available:", len(tf.config.list_physical_devices("GPU"))>0)

print("="*50)
print("Dataset Summary")
print("="*50)

print("Total interactions :", len(df))
print("Unique users :", df[colUser].nunique())
print("Unique places :", df[colPlace].nunique())
print("Rating range :", df[colRating].min(), "to", df[colRating].max())
print("Mean rating :", round(df[colRating].mean(),2))

print("="*50)

fig,axes = plt.subplots(1,3,figsize=(18,4))

axes[0].hist(df[colRating],bins=20)
axes[0].set_title('Rating Distribution')

axes[1].hist(df[colUser].value_counts(),bins=30)
axes[1].set_title('Ratings per User')

axes[2].hist(df[colPlace].value_counts(),bins=30)
axes[2].set_title('Ratings per Place')

plt.tight_layout()
plt.show()

catCounts = df[colCategory].value_counts().reset_index()
catCounts.columns=['Category','Count']

figCat = px.bar(catCounts,x='Category',y='Count',title='Places by Category')
figCat.show()

df = df.dropna(subset=[colUser,colPlace,colRating])

userEnc = LabelEncoder()
placeEnc = LabelEncoder()

df['userIdx'] = userEnc.fit_transform(df[colUser].astype(str))
df['placeIdx'] = placeEnc.fit_transform(df[colPlace].astype(str))

ratingMin = df[colRating].min()
ratingMax = df[colRating].max()

df['ratingNorm'] = (df[colRating]-ratingMin)/(ratingMax-ratingMin)

nUsers = df['userIdx'].nunique()
nPlaces = df['placeIdx'].nunique()

print("Users:",nUsers,"| Places:",nPlaces)

X = df[['userIdx','placeIdx']].values
y = df['ratingNorm'].values

xTemp,xTest,yTemp,yTest = train_test_split(X,y,test_size=testSplit,random_state=42)
xTrain,xVal,yTrain,yVal = train_test_split(xTemp,yTemp,test_size=validationSplit,random_state=42)

print("Train:",len(xTrain),"| Val:",len(xVal),"| Test:",len(xTest))

def buildModel(nUsers,nPlaces,embDim=64):

    userInput = layers.Input(shape=(1,))
    placeInput = layers.Input(shape=(1,))

    gmfU = layers.Flatten()(layers.Embedding(nUsers,embDim//2,
                    embeddings_regularizer=regularizers.l2(1e-6))(userInput))
    gmfP = layers.Flatten()(layers.Embedding(nPlaces,embDim//2,
                    embeddings_regularizer=regularizers.l2(1e-6))(placeInput))

    gmfOut = layers.Multiply()([gmfU,gmfP])

    mlpU = layers.Flatten()(layers.Embedding(nUsers,embDim,
                    embeddings_regularizer=regularizers.l2(1e-6))(userInput))
    mlpP = layers.Flatten()(layers.Embedding(nPlaces,embDim,
                    embeddings_regularizer=regularizers.l2(1e-6))(placeInput))

    x = layers.Concatenate()([mlpU,mlpP])

    x = layers.Dense(256,activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Dense(128,activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.25)(x)

    x = layers.Dense(64,activation='relu')(x)
    x = layers.Dropout(0.2)(x)

    mlpOut = layers.Dense(32,activation='relu')(x)

    fusion = layers.Concatenate()([gmfOut,mlpOut])
    output = layers.Dense(1,activation='sigmoid')(fusion)

    return Model(inputs=[userInput,placeInput],outputs=output)

model = buildModel(nUsers,nPlaces,embeddingDim)

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=learningRate),
    loss='mse',
    metrics=['mae']
)

model.summary()
```
